# **Gradiente Descendiente**

El descenso de gradiente (GD) es un algoritmo de optimización iterativo de primer orden, que se utiliza para encontrar un mínimo/máximo local de una función determinada. Se usa comúnmente en aprendizaje automático (ML) y aprendizaje profundo (DL) para minimizar una función de costo/pérdida (por ejemplo, en una regresión lineal).

image.png

 El descenso de gradiente actualiza iterativamente el valor de un parámetro (digamos θ ) dando un pequeño 'paso' en la dirección de la pendiente de la pérdida J( ∇J(θ​) ). La duración o el valor de este 'paso' está determinado por un hiperparámetro llamado tasa de aprendizaje (α). La longitud del 'paso' es α multiplicado por la pendiente ∇J (θ​) .

 *θnew​=θold​−α∇J(θold​)*

los pasos del método Gradiente Descendiente son:



1.   Elegir un punto de partida (inicialización)
2.   Calcular el gradiente en este punto
3.   Hacer un paso a escala en la dirección opuesta al gradiente (objetivo: minimizar)
4.   Repita los puntos 2 y 3 hasta que se cumpla uno de los criterios:
    *   Número máximo de iteraciones alcanzado
    *   El tamaño del paso es menor que la tolerancia (debido a la escala o a un pequeño gradiente).




```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

xold = 5  # Valor inicial de x
x = np.arange(-6, 6, 0.1)  # Rango de valores de x para graficar
f_x = x**2 + 5*np.sin(x)  # Función de costo a minimizar

plt.figure(figsize=(20, 10))  # Configura el tamaño de la figura para la gráfica
plt.plot(x, f_x)  # Grafica la función de costo

# Inicialización de variables
y = xold**2 + 5*np.sin(xold)

# Ciclo de optimización (descenso de gradiente)
for i in range(10):
    df = 2*xold + 5*np.cos(xold)  # Calcula el gradiente de la función en xold
    xnew = xold - df  # Actualiza x utilizando la tasa de aprendizaje
    y = xnew**2 + 5*np.sin(xnew)  # Calcula el valor de la función en el nuevo x
    plt.scatter(xnew, y, 200)  # Dibuja un punto en la gráfica en la nueva posición
    plt.annotate(i, (xnew, y))  # Añade una etiqueta con el número de iteración
    xold = xnew  # Actualiza xold con el nuevo valor

# Imprime el resultado final de la optimización
print("Resultado final: {0},{1}".format(xnew, y))

plt.show()  # Muestra la gráfica
```

## **Gradiente descendiente estocástico**

El tamaño del paso para la actualización de los valores de los parámetros es, quizás, el factor más crucial en Gradient Descent. Este tamaño lo decide un hiperparámetro llamado tasa de aprendizaje (representado como α).
Elegir la tasa de aprendizaje adecuada es el arte del aprendizaje automático. Si considera que la tasa de aprendizaje es demasiado pequeña, su modelo tardará mucho en converger a los mínimos.

Por otro lado, si la tasa de aprendizaje es demasiado grande, es posible que saltes los mínimos y llegues al otro lado del valle.

image.png

```python
lr = 0.1  # Tasa de aprendizaje
#lr = 0.1 #Probar con valores más grandes como este
xold = 5  # Valor inicial de x
x = np.arange(-6, 6, 0.1)  # Rango de valores de x para graficar
f_x = x**2 + 5*np.sin(x)  # Función de costo a minimizar
plt.figure(figsize=(10, 5))  # Configura el tamaño de la figura para la gráfica
plt.plot(x, f_x)  # Grafica la función de costo

# Inicialización de variables
y = xold**2 + 5*np.sin(xold)

# Ciclo de optimización (descenso de gradiente)
for i in range(40): #Probar variando la cantidad de epocas
    df = 2*xold + 5*np.cos(xold)  # Calcula el gradiente de la función en xold
    xnew = xold - lr * df  # Actualiza x utilizando la tasa de aprendizaje
    y = xnew**2 + 5*np.sin(xnew)  # Calcula el valor de la función en el nuevo x
    plt.scatter(xnew, y, 200)  # Dibuja un punto en la gráfica en la nueva posición
    plt.annotate(i, (xnew, y))  # Añade una etiqueta con el número de iteración
    xold = xnew  # Actualiza xold con el nuevo valor

# Imprime el resultado final de la optimización
print("Resultado final: {0},{1}".format(xnew, y))

plt.show()  # Muestra la gráfica
```

# **Momentum SGD**

Descenso de gradiente es un método muy robusto y se mantiene aproximandose constantemente hacia un mínimo local lo cual puede ocasionar:
*  En caso de tener funciones de costo con muchas pequeñas oscilaciones, GD es propenso a ser atrapado en “malos minímos locales”.
*  Si se esta muy lejos del óptimo y el tamaño de paso es pequeño, entonces GD tendrá una convergencia muy lenta

Por lo anterior, se incluye la inercia La idea de Momentum Optimization es acumular velocidad durante el descenso.

image.png

Para cada parámetro, v acumula la derivada parcial y ⍺ es un factor de impulso (es decir, 0,9) que decae gradualmente el efecto de las derivadas parciales pasadas.

```python
lr = 0.1  # Tasa de aprendizaje
alpha = 0.9 # Término de momento
xold = 5 # Valor inicial de x
x = np.arange(-6, 6, 0.1)  # Rango de valores de x para graficar
f_x = x**2 + 10*np.sin(x)  # Función de costo a minimizar
plt.figure(figsize=(20, 10))  # Configura el tamaño de la figura para la gráfica
plt.plot(x, f_x)  # Grafica la función de costo

# Inicialización de variables
y = xold**2 + 10*np.sin(xold)
v=0

# Ciclo de optimización (SGD con momento)
for i in range(50):
    df = 2*xold + 10*np.cos(xold)  # Calcula el gradiente de la función en xold

    # Aplica el término de momento y ajusta los valores
    #xnew = xold - lr*(gamma + df)
    v = alpha* v + lr * df
    xnew = xold - v

    y = xnew**2 + 10*np.sin(xnew)  # Calcula el valor de la función en el nuevo x

    # Dibuja un punto en la gráfica en la nueva posición
    plt.scatter(xnew, y, 200)

    # Añade una etiqueta con el número de iteración
    plt.annotate(i, (xnew, y))

    xold = xnew  # Actualiza xold con el nuevo valor

# Muestra la gráfica
# plt.show()
```

## **Nesterov (NAG)**

Una variante del impulso SGD, el impulso de Nesterov es una técnica que realiza una actualización más informada. En Nesterov, se calcula el gradiente no en la posición actual, sino en una posición "futura" que es una estimación de dónde podría estar el próximo paso con el término de momento. Luego, se utiliza este gradiente estimado para actualizar la posición actual.


La adición de este pequeño término ayuda a que la actualización siempre se acerque más al óptimo. Esto ayuda de dos maneras:


1.   Aumenta la velocidad de convergencia.
2.   Reduce las oscilaciones alrededor del óptimo.


image.png

image.png

Por lo tanto, el impulso se actualiza con el gradiente en una posición de anticipación, incorporando el valor del gradiente futuro en la actualización del parámetro actual.

Si los gradientes se hacen más pequeños, el impulso de Nesterov reaccionaría más rápido que el algoritmo de impulso original.

```python
lr = 0.1  # Tasa de aprendizaje
alpha = 0.9 # Término de momento
xold = 5 # Valor inicial de x
x = np.arange(-6, 6, 0.1)  # Rango de valores de x para graficar
f_x = x**2 + 10*np.sin(x)  # Función de costo a minimizar
plt.figure(figsize=(20, 10))  # Configura el tamaño de la figura para la gráfica
plt.plot(x, f_x)  # Grafica la función de costo

# Inicialización de variables
y = xold**2 + 10*np.sin(xold)
v=0

# Ciclo de optimización (SGD con momento)
for i in range(50):
    df = 2*(xold - alpha * v) + 10*np.cos(xold - alpha * v) # Calcula el gradiente de la función en xold

    # Aplica el término de momento y ajusta los valores
    #xnew = xold - lr*(gamma + df)
    v = alpha * v + lr * df
    xnew = xold - v

    y = xnew**2 + 10*np.sin(xnew)  # Calcula el valor de la función en el nuevo x

    # Dibuja un punto en la gráfica en la nueva posición
    plt.scatter(xnew, y, 200)

    # Añade una etiqueta con el número de iteración
    plt.annotate(i, (xnew, y))

    xold = xnew  # Actualiza xold con el nuevo valor

# Muestra la gráfica
# plt.show()
```
